```python
x = np.random.rand(50)
y = np.random.rand(50)

x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
colors = np.random.rand(50)
plt.scatter(x,y, c=colors)
```

```python
sizes = 1000*np.random.rand(50)
colors = np.random.rand(50)
plt.scatter(x,y, c=colors, s= sizes)
```

```python
sizes = 1000*np.random.rand(50)
colors = np.random.rand(50)
plt.scatter(x,y, c=colors, s=sizes, alpha=.1)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Random Data")
```
