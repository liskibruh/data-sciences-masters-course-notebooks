# Seaborn

```python
import seaborn as sns
```

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

```python
sns.scatterplot(x = iris['sepal_length'], y= iris['sepal_width'])
```

```python
sns.scatterplot(x = iris['sepal_length'], y= iris['petal_length'])
```

```python
sns.scatterplot(x = iris['petal_width'], y= iris['petal_length'])
```

```python
sns.displot(iris['sepal_length'])
```

```python
tips= sns.load_dataset('tips')
```

```python
tips
```

```python
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

```python
tips.head()
```

```python
tips['smoker'].value_counts()
```

```python
sns.relplot(x=tips['total_bill'], y=tips['tip'], data=tips, hue='smoker')
```

```python
tips.head()
```

```python
sns.relplot(x=tips['total_bill'], y=tips['tip'], data=tips, hue='size')
```

```python
sns.relplot(x=tips['total_bill'], y=tips['tip'], data=tips, style='size')
```

```python
sns.relplot(x=tips['total_bill'], y=tips['tip'], data=tips,hue='time', style='size')
```

```python
tips.head()
```

```python
sns.catplot(data=tips, x=tips['day'], y=tips['total_bill'])
```

```python
sns.catplot(data=tips, x=tips['smoker'], y=tips['tip'])
```

```python
sns.jointplot(x=tips.total_bill, y=tips.tip)
```

```python
sns.pairplot(tips)
```

```python
sns.pairplot(iris)
```

```python
sns.jointplot(iris)
```
