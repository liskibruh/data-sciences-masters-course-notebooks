```python
import plotly.graph_objects as go
import seaborn as sns
```

# Question No. 1:
Load the "titanic" dataset using the load_dataset function of seaborn. Use Plotly express to plot a
scatter plot for age and fare columns in the titanic dataset.

## Answer:

```python
titanic=sns.load_dataset("titanic")
titanic.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=titanic.age, y=titanic.fare, mode='markers'))
fig.show()
```

# Question No. 2:
Using the tips dataset in the Plotly library, plot a box plot using Plotly express.

## Answer:

```python
tips= sns.load_dataset("tips")
tips.head()
```

```python
fig = go.Figure()
fig.add_trace(go.Box(y=tips.total_bill))
fig.show()
```

# Question No. 3:
Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
column with the color parameter.

## Answer:

```python
import plotly.express as px
fig = px.bar(tips, x="sex", y="total_bill", color="day", pattern_shape="smoker")
fig.show()
```

# Question No. 4:
Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for

# Question No. 5:
What is Distplot? Using Plotly express, plot a distplot.

## Answer:
**Distplot:** A Distplot or distribution plot, depicts the variation in the data distribution. <br>
The distplot function in Python creates a histogram of the data, which shows the frequency of different values or ranges of values in the dataset.

```python
import plotly.express as px
tips = sns.load_dataset("tips")
fig = px.histogram(tips, x="total_bill", y="tip", color="sex", marginal="rug",
                   hover_data=tips.columns)
fig.show()
```
