# Question No. 1:
Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

## Answer:
1. **scatterplot():** Seaborn scatterplot is a visualization tool used to explore the relationship between two variables. It is a type of plot that displays the values of two variables as points on a two-dimensional coordinate system.<br>The primary use of a seaborn scatterplot is to identify patterns or trends in the data. It helps in understanding the relationship between two variables, such as whether they are positively or negatively correlated.<br>In addition to exploring relationships between two variables, seaborn scatterplots can also be used to compare multiple groups of data. By adding color or size to the points on the scatterplot, different groups of data can be distinguished and compared. 

```python
import seaborn as sns

tips=sns.load_dataset("tips")
sns.scatterplot(x=tips['total_bill'], y=tips['tip'])
```

2. **displot():** The primary use of seaborn displot() is to visualize the distribution of a dataset and to identify any underlying patterns or trends. It helps to explore the shape of the data distribution, such as whether it is symmetric or skewed, and to identify any outliers or unusual values.

```python
iris = sns.load_dataset("iris")
sns.displot(iris['sepal_length'])
```

3. **pairplot():** Pairplot displays scatterplots between all possible pairs of variables in a dataset, and can also include histograms or kernel density plots of the variables along the diagonal of the plot. This allows for a quick overview of the relationships between variables in a dataset.<br>The primary use of seaborn pairplot is to identify patterns or trends between pairs of variables in a dataset. It can be used to explore the correlation between variables, to identify potential outliers or data errors, and to visualize the distribution of the data.

```python
sns.pairplot(iris)
```

4. **jointplot():** Jointplot displays a scatterplot between the two variables in the center of the plot, along with the histograms or kernel density plots of each variable along the x and y axes. This allows for a quick overview of the relationship between the two variables, as well as their individual distributions.<br>The primary use of seaborn jointplot is to identify patterns or trends between two variables in a dataset. It can help visualize the correlation between the two variables, as well as any potential outliers or data errors.

```python
sns.jointplot(x=tips.total_bill, y=tips.tip)
```

5. **lineplot():** The lineplot() function is similar to the plot() function in Matplotlib but provides additional functionality for uncertainty estimation. It can plot a line connecting the points in the dataset, as well as display confidence intervals or error bars to indicate the uncertainty in the data.<br>The primary use of lineplot() is to identify trends or patterns in the data over time or across a continuous variable. It can also help identify outliers or data errors that may be affecting the overall patterns in the data.

```python
sns.lineplot(x=tips.total_bill, y=tips.tip)
```

# Question No. 4:
Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
parameter for the 'cut' column of the diamonds dataset.

## Answer:

```python
diamonds = sns.load_dataset("diamonds")
diamonds.head()
```

```python
sns.histplot(data=diamonds, x=diamonds.price, hue='cut')
```

# Question No. 6:
Use the "flights" dataset from seaborn to plot a heatmap.

## Answer:

```python
flights = sns.load_dataset("flights")
flights.head()
```

```python
flights_pv = flights.pivot('month', 'year', 'passengers')
flights_pv
```

```python
sns.heatmap(flights_pv)
```
