# Question No. 13:
How does the type of app affect its price? Create a bar chart to compare average prices by app type.

## Answer:

```python
df=df[df['Type']!='0']
```

```python
df.Type.value_counts()
```

```python
chars_to_remove = ['+',',','$']
cols_to_clean = ['Installs','Price']
for item in chars_to_remove:
    for cols in cols_to_clean:
        df[cols]=df[cols].str.replace(item,'')
```

```python
df['Price']=df['Price'].astype(float)
```

```python
avg_price_by_type = df.groupby('Type')['Price'].mean().reset_index()

# Create a bar chart using Seaborn
sns.barplot(x='Price', y='Type', data=avg_price_by_type)
```

- **We can clearly see that the free apps has no price, and the paid apps has an average price of 14$**
